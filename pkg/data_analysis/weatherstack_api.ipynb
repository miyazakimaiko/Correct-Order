```python
from datetime import datetime
```

```python
hourlyweather = []
for index in hourly.index:
    for column in hourly.columns:
        new_df = hourly.loc[index, column]
        dt = datetime.strptime(dates.loc[index, 'date'], '%Y-%m-%d')
        date = dt.strftime('%m/%d/%Y')
        hour = int(int(new_df['time'])/100)
        new_df['MM/DD/YY'] = date
        new_df['datetime'] = '{} {}:00:00'.format(date, hour)
        new_df['hour'] = hour
        hourlyweather.append(new_df)
```

```python
hourlyweather = pd.DataFrame(hourlyweather)
hourlyweather
```

```python
df_hourly = hourlyweather.drop(['chanceofremdry', 'chanceofwindy', 'chanceofovercast', 'chanceofsunshine',
                                'chanceoffrost', 'chanceofhightemp', 'chanceoffog', 'chanceofsnow', 'cloudcover', 
                                'chanceofthunder', 'uv_index', 'wind_degree', 'wind_dir', 'weather_icons',
                                'humidity', 'visibility', 'pressure', 'heatindex', 'dewpoint', 'windchill',
                                'windgust', 'feelslike', 'chanceofrain', 'weather_descriptions'], axis=1)
```

```python
df_hourly
```

```python
for index in df_hourly.index:
    if df_hourly.loc[index, 'hour'] == 0 or df_hourly.loc[index, 'hour'] == 18:
        df_hourly = df_hourly.drop([index])
    else:
        date = df_hourly.loc[index, 'datetime']
        df_hourly = df_hourly.rename(index={index: date})
```

```python
df_hourly = df_hourly.drop(['time', 'datetime'], axis=1)
```

```python
df_hourly
```

```python
from decimal import *
```

```python
average = {}
temp = None
wind = None
rain = None
new_index = None
for index in df_hourly.index:
    if df_hourly.loc[index, 'hour'] == 6:
        new_index = df_hourly.loc[index, 'MM/DD/YY']
        average[new_index] = {}
        average[new_index]['weather_morning'] = df_hourly.loc[index, 'weather_code']

        temp = df_hourly.loc[index, 'temperature']
        wind = df_hourly.loc[index, 'wind_speed']
        rain = df_hourly.loc[index, 'precip']
        
    elif df_hourly.loc[index, 'hour'] == 12:
        average[new_index]['weather_afternoon'] = df_hourly.loc[index, 'weather_code']
        average[new_index]['temperature'] = (temp + df_hourly.loc[index, 'temperature']) / 2
        average[new_index]['wind_speed'] = (wind + df_hourly.loc[index, 'wind_speed']) / 2
        rain = (Decimal(int(rain) + Decimal(int(df_hourly.loc[index, 'precip'])))) / 2
        average[new_index]['precip'] = format(rain, '.1f')

        temp = None
        wind = None
        rain = None
        new_index = None
```

```python
average = pd.DataFrame(average)
```

```python
average = pd.DataFrame.transpose(average)
```
