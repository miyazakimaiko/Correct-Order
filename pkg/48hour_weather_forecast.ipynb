```python
import requests
import pandas as pd
```

```python
url = 'https://api.openweathermap.org/data/2.5/onecall?lat=53.349805&lon=-6.26031&exclude=current,minutely,daily,alerts&appid=4fab58f885a6f341a28007b52346cffc'
result = requests.post(url=url)
```

```python
json_result = result.json()
```

```python
df = pd.DataFrame(json_result)
```

```python
df
```

```python
hourly = pd.DataFrame([md for md in df.hourly])
hourly
```

```python
weather = pd.DataFrame([md for md in hourly.weather])
```

```python
weather = pd.DataFrame([md for md in weather[0]])
```

```python
merged_df = hourly.merge(weather, left_index=True, right_index=True)
merged_df
```

```python
del merged_df['feels_like']
del merged_df['pressure']
del merged_df['humidity']
del merged_df['dew_point']
del merged_df['visibility']
del merged_df['wind_deg']
del merged_df['weather']
del merged_df['pop']
del merged_df['rain']
del merged_df['main']
del merged_df['description']
del merged_df['icon']
```

```python
from datetime import datetime
```

```python
for index in merged_df.index:
    if merged_df.loc[index, 'dt']:
        t = merged_df.loc[index, 'dt']
        merged_df.loc[index, 'dt_iso'] = datetime.utcfromtimestamp(t).strftime('%Y-%m-%d %H:%M:%S')
```

```python
merged_df['dt_iso'] = pd.to_datetime(merged_df['dt_iso'], errors='coerce')
merged_df['hour'] = merged_df['dt_iso'].dt.hour
merged_df
```

```python
merged_df = merged_df.loc[(merged_df['hour']==9) | (merged_df['hour']==13) | (merged_df['hour']==17)]
merged_df = merged_df.rename(columns={'id':'weather_id'})
merged_df = merged_df.reset_index(drop=True)
```

```python
merged_df
```

```python
for index in merged_df.index:
    if merged_df.loc[index, 'dt_iso']:
        t = merged_df.loc[index, 'dt_iso'].strftime('%m/%d/%Y  %H:%M:%S')
        merged_df.loc[index, 'dt_iso'] = t

merged_df
```
