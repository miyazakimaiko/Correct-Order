```python
import requests
import json
import pandas as pd
```

```python
url = 'https://api.weatherstack.com/historical'
api_key = 'secret'
startdate = '2019-01-01'
enddate = '2019-01-31'
params = {
    'access_key': api_key,
    'query': 'Dublin',
    'historical_date_start': startdate,
    'historical_date_end': enddate,
    'hourly': 1,
    'interval': 1,
    'units': 'm'
}
```

```python
result = requests.post(url=url, params=params)
```

```python
result_json = result.json()
df = pd.DataFrame(result_json)
```

```python
df = df.drop(['type', 'query', 'language', 'unit', 'name', 'country', 
             'region', 'lat', 'lon', 'timezone_id', 'localtime',
             'localtime_epoch', 'utc_offset', 'observation_time',
             'temperature', 'weather_code', 'weather_icons', 
             'weather_descriptions', 'wind_speed', 'wind_degree',
             'wind_dir', 'pressure', 'precip', 'humidity', 'cloudcover',
             'feelslike', 'uv_index', 'visibility', 'is_day'])

df = df.drop(['request', 'location', 'current'], axis=1)
```

```python
df
```

```python
daily = pd.DataFrame([md for md in df.historical])
daily
```

```python
dates = daily.drop(['date_epoch', 'astro', 'mintemp', 'maxtemp', 'avgtemp', 
                    'totalsnow', 'sunhour', 'uv_index', 'hourly'], axis=1)
dates
```

```python
hourly = pd.DataFrame([md for md in daily.hourly])
hourly
```

```python
hourly = hourly.drop([0, 1, 2, 3, 4, 5, 6, 18, 19, 20, 21, 22, 23], axis=1)
```

```python
from datetime import datetime
```

```python
hourlyweather = []
for index in hourly.index:
    for column in hourly.columns:
        new_df = hourly.loc[index, column]
        dt = datetime.strptime(dates.loc[index, 'date'], '%Y-%m-%d')
        date = dt.strftime('%m/%d/%Y')
        hour = int(int(new_df['time'])/100)
        new_df['MM/DD/YY'] = date
        new_df['datetime'] = '{} {}:00:00'.format(date, hour)
        new_df['weeknum'] = dt.strftime("%W")
        new_df['dayofweek'] = pd.Timestamp(pd.to_datetime(date)).dayofweek
        new_df['hour'] = hour
        new_df['date'] = pd.to_datetime(date).day
        new_df['month'] = pd.to_datetime(date).month
        hourlyweather.append(new_df)
```

```python
hourlyweather = pd.DataFrame(hourlyweather)
hourlyweather
```

```python
df_hourly = hourlyweather.drop(['chanceofremdry', 'chanceofwindy', 'chanceofovercast', 'chanceofsunshine',
                                'chanceoffrost', 'chanceofhightemp', 'chanceoffog', 'chanceofsnow', 
                                'chanceofthunder', 'uv_index', 'wind_degree', 'wind_dir', 'weather_icons',
                                'humidity', 'visibility', 'pressure', 'heatindex', 'dewpoint', 'windchill',
                                'windgust', 'feelslike', 'chanceofrain', 'weather_descriptions'], axis=1)
```

```python
df_hourly
```

```python
for index in df_hourly.index:
    date = df_hourly.loc[index, 'datetime']
    df_hourly = df_hourly.rename(index={index: date})
```

```python
df_hourly = df_hourly.drop(['time', 'datetime'], axis=1)
```
