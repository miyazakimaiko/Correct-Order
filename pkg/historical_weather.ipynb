```python
import pandas as pd
import os
import sys
```

```python
df = pd.read_json(r'./data/weather01012018to30092020.json')
df
```

```python
df = pd.DataFrame(df)
df
```

```python
for index in df.index:
    if df.loc[index,'dt_iso']:
        new = df.loc[index,'dt_iso'].replace('+0000 UTC', '')
        df.loc[index,'dt_iso'] = new
```

```python
df['dt_iso'] = pd.to_datetime(df['dt_iso'], errors='coerce')
df['hour'] = df['dt_iso'].dt.hour
df['year'] = df['dt_iso'].dt.year
df['month'] = df['dt_iso'].dt.month

df.head()
```

```python
new_df = df.loc[(df['year']==2020) & (df['month']==7)]
new_df = new_df.loc[(df['hour']==7) | (df['hour']==14)| (df['hour']==15)| (df['hour']==16)| (df['hour']==17)]
```

```python
new_df
```

```python
main = pd.DataFrame([md for md in new_df.main])
del main['temp_min']
del main['temp_max']
del main['feels_like']
del main['pressure']
del main['humidity']

main
```

```python
wind = pd.DataFrame([md for md in new_df.wind])
wind = wind.rename(columns={'speed':'wind_speed'})
del wind['deg']
wind
```

```python
clouds = pd.DataFrame([md for md in new_df.clouds])
clouds = clouds.rename(columns={'all':'clouds'})
clouds
```

```python
weather = pd.DataFrame([md for md in new_df.weather])
weather = pd.DataFrame([md for md in weather[0]])
weather = weather.rename(columns={'id':'weather_id'})

del weather['main']
del weather['icon']

weather
```

```python
del new_df['city_name']
del new_df['lat']
del new_df['lon']
del new_df['main']
del new_df['wind']
del new_df['rain']
del new_df['clouds']
del new_df['weather']
del new_df['timezone']
del new_df['snow']
```

```python
new_df = new_df.reset_index(drop=True)
```

```python
new_df
```

```python
merged_df = pd.concat([new_df, main, wind, clouds, weather], axis=1, join='inner')
merged_df
```

```python
for index in merged_df.index:
    if merged_df.loc[index, 'dt_iso']:
        t = merged_df.loc[index, 'dt_iso'].strftime('%m/%d/%Y  %H:%M:%S')
        merged_df.loc[index, 'dt_iso'] = t

merged_df
```
